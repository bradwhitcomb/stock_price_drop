```python
import pickle

for stock in stocks:
    stock = Investment(stock)
    price = stock.grab_price
    symbol = stock.symbol
    hw_dict = {}
    delta_dict = {}
    upperSymbol = symbol.upper()
    
    try:
        
        with open(f"{symbol}.pickle","rb") as pickle_off:
            highwater = pickle.load(pickle_off)
#     except:
#         print("No file!  Let's create one")
        
    except:
        print(f"No file for {symbol}!  Let's create one")
        with open(f"{symbol}.pickle","wb") as pickling_on:
                pickle.dump(price, pickling_on)
                highwater = price
                
#     except:
#         print("We have a problem creating a new file")
        
        
    if price > highwater:
    
        with open(f"{symbol}.pickle","wb") as pickling_on:
            pickle.dump(price, pickling_on)
                
            
            
            
            print(f"The previous close for {stock.symbol} is ${price:,.2f}. The highwater mark is ${highwater:,.2f}")
            print(" ")
    else:   
        print(f"The previous close for {stock.symbol} is ${price:,.2f}. The highwater mark is ${highwater:,.2f}")
        print(f"Last price is {(highwater - price)/highwater:.2%} below the highwater mark")
        print(" ")
            
    hw_dict['HWater'] = highwater
    investment[upperSymbol].update( hw_dict )
    delta = (price - highwater) / price
    delta_dict['% change'] = delta
    investment[upperSymbol].update(delta_dict)
    
```
